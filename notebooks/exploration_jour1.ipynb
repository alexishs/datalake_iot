## 7. Format & paramètres d'import

**Lecture CSV** — séparateur `,` · décimale `.` · encodage **UTF-8** · en-tête **présent** (1 ligne) · 1 ligne = 1 relevé.

**Colonnes & unités** *(sources : `data/How_to_Use.pdf` et `Synthetic Data from Industrial Sensor.pdf`)*

| colonne (cible) | type | unité | notes |
|---|---|---|---|
| `timestamp` | datetime | — | format source `YYYY-MM-DD HH:MM:SS` (espace, **non ISO-T**), **sans fuseau** ; pas = 1 min |
| `temperature` | float | **°C** | casse source variable (`Temperature`/`temperature`) |
| `pressure` | float | **bar** | casse source variable (`Pressure`/`pressure`) |
| `elapsed_time` | float | arbitraire | **temps de fonctionnement machine** (doc) ; **optionnel** (lignes A/B) ; casse variable |
| `label` | entier {0,1} | — | cible anomalie (0=nominal, 1=anomalie) ; nommé `anomaly` dans la doc |

**Identifiant de ligne** : motif `Line([A-E])` dans le nom de fichier → `lineA`…`lineE`.

## 8. Schéma cible unifié (`staging/`)

Schéma de sortie figé après harmonisation :

| colonne | type | nullable | provenance |
|---|---|---|---|
| `timestamp` | datetime (ISO 8601, UTC) | non | parsé depuis la source |
| `temperature` | float (°C) | non | colonne renommée en minuscules |
| `pressure` | float (bar) | non | colonne renommée en minuscules |
| `elapsed_time` | float | **oui** | absente de C/D/E → `NaN` |
| `label` | entier {0,1} | non | inchangée |
| `line` | catégorie | non | dérivée du nom de fichier |
| `year`, `month` | entier / texte | non | colonnes de partition (extraites du `timestamp`) |


- **Fuseau horaire** du `timestamp` : non précisé → sera considéré comme **UTC** .
- `elapsed_time` : « temps de fonctionnement machine », unité arbitraire, d'après la description du jeu de données.

## Synthèse — règles techniques pour l'implémentation de l'import

Liste **exhaustive** des règles que le script/DAG d'ingestion (`raw/`) et
d'harmonisation (`staging/`) doit respecter. Les faits détaillés sont en §1–§8.

### Lecture des fichiers (à figer — ne rien laisser deviner)
1. CSV : séparateur `,`, décimale `.`, encodage **UTF-8**, en-tête présent (1 ligne). Fixer ces paramètres explicitement.
2. **Ne pas se fier à l'ordre des colonnes** (il varie) : adresser chaque colonne par son **nom normalisé**.
3. **Ne pas se fier au nombre de colonnes** (4 ou 5 selon la ligne).

### Harmonisation des colonnes (→ `staging/`)
4. **Normaliser tous les noms de colonnes en minuscules** (résout `Temperature`/`temperature`, `Pressure`/`pressure`, `Elapsed_time`/`elapsed_time`).
5. **`elapsed_time` optionnel** : si absent (LineC/D/E), créer la colonne avec `NaN`. Ne jamais présumer sa présence.
6. Produire le **schéma cible figé du §8** (mêmes colonnes, mêmes types) quel que soit le fichier d'entrée.

### Typage / parsing
7. `timestamp` : parser explicitement le format `%Y-%m-%d %H:%M:%S` → datetime, **normaliser en ISO 8601** ; fuseau **supposé UTC** (à documenter).
8. `temperature`, `pressure`, `elapsed_time` → `float` ; `label` → entier `{0, 1}`.

### Partitionnement — ⚠️ règle cruciale
9. Chemin de dépôt : `raw/production_lines/{line}/year={YYYY}/month={MM}/…`.
10. `line` (`lineA`…`lineE`) se déduit du **nom de fichier** (motif `Line([A-E])`) — légitime : un fichier = une ligne de production.
11. **`year` / `month` se déduisent du `timestamp` de CHAQUE enregistrement**, jamais du nom de fichier ni d'une quelconque hypothèse.
12. **Aucune coïncidence fichier ↔ partition ne doit être exploitée.** Bien que, dans ce jeu de données, chaque fichier tienne dans un seul mois, **le code ne doit en aucun cas s'appuyer là-dessus** : il doit **grouper les enregistrements par `(year, month)` issus du `timestamp`** et écrire dans **autant de partitions que nécessaire**. Un fichier qui chevaucherait deux mois (ou plus) doit être réparti automatiquement sur les bonnes partitions, **sans modification du code**.

### Intégrité
13. Vérifier le **MD5** des fichiers sources (fait au téléchargement) et contrôler l'intégrité **après dépôt** dans `raw/` (comparaison MD5/ETag).

### Ingestion par chunks
14. Traiter les fichiers **par chunks** (obligatoire pour LineA — 10 000 lignes ; appliqué par principe à tous) pour simuler un flux.
15. Le découpage en chunks **ne doit pas** dicter le partitionnement : regrouper par `(year, month)` **indépendamment** des frontières de chunks (un même mois peut s'étaler sur plusieurs chunks, et inversement).

### Idempotence & reproductibilité
16. L'ingestion doit être **relançable sans créer de doublons** (clés d'objets déterministes, écrasement contrôlé).
