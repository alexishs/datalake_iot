# Exploration des CSV bruts

## 1. Chargement

```python
# Recharge auto du package `datalake` à chaque exécution (dev sans redémarrage)
%load_ext autoreload
%autoreload 2

import polars as pl
from datalake import explore

pl.Config.set_tbl_cols(50)         # afficher toutes les colonnes
pl.Config.set_fmt_str_lengths(80)  # ne pas tronquer les chaînes longues
paths = explore.csv_paths()
[p.name for p in paths]
```

## 1 bis. Aperçu — 5 premières lignes de chaque fichier

Pour visualiser concrètement le contenu : valeurs, format du `timestamp`, et la
casse réelle des colonnes telle qu'elle apparaît dans chaque fichier.

```python
from IPython.display import display, Markdown

for p in paths:
    display(Markdown(f"**{p.name}** — aperçu (5 lignes)"))
    display(explore.head(p))
    display(Markdown("_Statistiques descriptives_"))
    display(explore.describe(p))
```

## 2. Volumétrie

```python
explore.volumetry(paths)
```

### Observations

- **LineA** : 10 000 enregistrements, soit **2× les autres** (5 000) → justifie le
  traitement **par chunks** demandé (simulation d'un flux).
- Nombre de colonnes : **5** pour LineA/LineB, **4** pour LineC/D/E (cf. `elapsed_time`).
- **0 valeur manquante** sur l'ensemble des fichiers.

## 3. Comparaison des schémas (casse des colonnes)

```python
explore.casing_report(paths)
```

### Observations

Hétérogénéités à harmoniser en `staging/` :

- **Casse non uniforme** : `temperature`/`Temperature`, `pressure`/`Pressure`
  (et `elapsed_time`/`Elapsed_time`) selon les fichiers → **normaliser en minuscules**.
- **`elapsed_time` optionnel** : présent dans LineA et LineB seulement, absent de
  LineC/D/E → prévoir une colonne **facultative** (valeur manquante = **`null`** natif Polars, persistée en `NULL` Parquet).
- `timestamp` et `label` sont cohérents (nom + casse) sur les 5 lignes.

## 4. Types de données

```python
explore.dtypes_table(paths)
```

## 5. `label` — valeurs & distribution

```python
explore.label_distribution(paths)
```

### Observations — `label`

- `label` vaut **{0, 1}** partout : `0` = fonctionnement nominal, `1` = anomalie.
  C'est la cible du futur projet de maintenance prédictive.

### Écart doc ↔ données (à tracer pour C20)

Le « % of Anomalies » annoncé par la description du jeu de données **ne correspond
pas toujours** au comptage réel ci-dessus :

| ligne | annoncé (doc) | mesuré (données) |
|---|---|---|
| LineE | **0 %** | **0,5 %** (25 anomalies) |
| LineB | « < 1 % » | **1,00 %** (pile) |

→ **La donnée fait foi.** À consigner comme avertissement qualité dans la fiche
OpenMetadata : ne pas présumer LineE « sans anomalie ».

## 6. Couverture temporelle & partitionnement

Période, cadence et continuité par fichier — c'est ce qui détermine les partitions
`year=YYYY/month=MM/` et l'identifiant `lineX` du chemin de dépôt en `raw/`.

```python
explore.coverage(paths)
```

### Observations — couverture

- **Échantillonnage régulier : 1 relevé/minute, continu** (ni trou ni doublon).
- **Un fichier = un seul mois** → une seule partition par ligne :

| ligne | chemin `raw/production_lines/…` |
|---|---|
| lineE | `lineE/year=2025/month=01/` |
| lineD | `lineD/year=2025/month=02/` |
| lineC | `lineC/year=2025/month=03/` |
| lineB | `lineB/year=2025/month=04/` |
| lineA | `lineA/year=2025/month=05/` |


## 7. Format observé (constats)

- **CSV** : séparateur `,`, en-tête présent (1 ligne), 1 ligne = 1 relevé.
- **`timestamp`** : format `YYYY-MM-DD HH:MM:SS` (espace, **non ISO-T**), **sans fuseau** ; cadence 1 min.
- **Unités** (d'après `data/How_to_Use.pdf`) : `temperature` en **°C**, `pressure` en **bar** ;
  `elapsed_time` = **temps de fonctionnement machine**, **unité arbitraire** (description du jeu de données).

→ Le **schéma cible**, les **paramètres d'import** et les **règles (contrat)** sont consignés dans
**[docs/architecture.md](../docs/architecture.md)** (§5–§6 et §12) — source de référence pour les DAGs.

L'exploration confirme les hétérogénéités annoncées et en quantifie d'autres :

- **Casse des colonnes** non uniforme (et parfois incohérente au sein d'un même fichier) ;
  `elapsed_time` présent uniquement sur LineA et LineB.
- **Volumétrie** : LineA = 10 000 lignes (2× les autres), **1 relevé/minute, continu** ;
  un **mois distinct par ligne** (janvier→mai 2025).
- **`label`** binaire `{0,1}`, fortement **déséquilibré** (~1 %), avec un **écart doc↔données**
  sur LineE (annoncé 0 %, mesuré 0,5 %).
- **Unités** : `temperature` en °C, `pressure` en bar ; `elapsed_time` = temps de
